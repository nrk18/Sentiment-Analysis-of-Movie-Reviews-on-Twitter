# Scraping Data from Twitter

```python
# importing libraries and packages
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data 
tweets_movie_review = []

#Setting search words that relates to tweets about veganism 
key= ['#movie review', '#moviereview','#filmreview','#film review']

# Using TwitterSearchScraper to scrape data and append tweets to list
for keyword in key:
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper
                             (f'{keyword} lang:en')
                             .get_items()): #declaring the search word, language of tweet for which tweets are extracted
        if i>50000: #setting number of tweets to scrape
            break
        tweets_movie_review.append([tweet.date, tweet.id, 
                             tweet.content, tweet.user.username, 
                             tweet.coordinates]) #declare the attributes to be returned
   
        
# Creating a dataframe from the tweets list above 
tweets_df = pd.DataFrame(tweets_movie_review, columns=['Datetime', 'Tweet Id', 'Text', 'Username','Location'])
```

# Saving the Scrapped Data into CSV file

```python
# Saving Data frame to a csv file

tweets_df.to_csv('movie_reviews_tweets.csv')
```
